#Calculando os melhores Features para o modelo

O método SelectKBest com o score de informação mútua clássica (mutual info classic) é a ferramenta mais utilizada na seleção de características (feature selection). O Mutual Info Classic quantifica a dependência mútua entre duas variáveis. No contexto da seleção de características, ele é usado para avaliar a relação entre cada característica (variável independente) e a variável alvo (variável dependente). Ele mede o quanto de informação a presença ou ausência de uma determinada característica pode fornecer sobre a variável alvo.

#Criando o modelo RandomForest de Regressão

utilizar dados aleatorios pra verificar a previsao deste metodo e como testar a acuracia

```python
##Treino e previsao, separando uma parte dos dados orginais pra fazer a previsao e apos testar a acuracia

# Definir as variáveis de entrada (X) e a variável de saída (y)
X = df[['non_solus', 'number_of_games', 'clashofclans_rank', 'royaleclash_rank',
       'clashof_lords2_rank', 'royaleclans_sow', 'plantsvszombies_sow',
       'lorsmobile_sow', 'clashoflords2_sow', 'other_sow']]  #variáveis escolhidas
y = df['clashofclans_sow']  #coluna alvo

# Dividir o conjunto de dados em treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Criar o modelo de Random Forest
model = RandomForestRegressor()

# Treinar o modelo com os dados de treinamento
model.fit(X_train, y_train)

# Fazer previsões usando o modelo treinado nos dados de teste
y_pred = model.predict(X_test)

# Calcular a acurácia do modelo usando a métrica R²
r2 = r2_score(y_test, y_pred)

# Imprimir a acurácia do modelo
print('Acurácia:', r2)
```

#Validação

####Coeficiente de Determinação (R²): Mede a proporção da variância da variável de saída que é explicada pelo modelo. Ele varia entre 0 e 1, onde 1 indica um ajuste perfeito do modelo aos dados. Quanto maior o valor de R², melhor é o ajuste do modelo.
####Erro Médio Absoluto (MAE): Média das diferenças absolutas entre as previsões do modelo e os valores reais da variável de saída. Ele fornece uma medida da média do erro absoluto cometido pelo modelo. Quanto menor o valor do MAE, melhor é o ajuste do modelo.
####Erro Quadrático Médio (MSE): Média dos erros ao quadrado entre as previsões do modelo e os valores reais da variável de saída. Ele fornece uma medida do erro médio quadrático cometido pelo modelo. Quanto menor o valor do MSE, melhor é o ajuste do modelo.
####Raiz do Erro Quadrático Médio (RMSE):A raiz quadrada do MSE e fornece uma medida do erro médio quadrático em escala original.


```python

# Calcular o R²
r2 = r2_score(y_test, y_pred)

# Calcular o MAE
mae = mean_absolute_error(y_test, y_pred)

# Calcular o MSE
mse = mean_squared_error(y_test, y_pred)

# Calcular o RMSE
rmse = mean_squared_error(y_test, y_pred, squared=False)

# Imprimir as métricas
print('R²:', r2)
print('MAE:', mae)
print('MSE:', mse)
print('RMSE:', rmse)
```
